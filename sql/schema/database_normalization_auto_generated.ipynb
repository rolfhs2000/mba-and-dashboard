```python
# This code add "DROP TABLE IF EXIST..." from file `*_raw.sql` and create new file `*.sql`
import re

input_path = '../../sql/schema/database_denormalization_raw.sql'
output_path = '../../sql/schema/database_denormalization.sql'

with open(input_path, encoding='utf-8') as f:
    ddl = f.read()

# Thêm lệnh DROP TABLE IF EXISTS trước CREATE TABLE
ddl = re.sub(
    r'(CREATE TABLE\s+`?(\w+)`?\s*\()',
    lambda m: f'DROP TABLE IF EXISTS `{m.group(2)}`;\n{m.group(1)}',
    ddl
)

with open(output_path, 'w', encoding='utf-8') as f:
    f.write(ddl)

print(f'Đã tạo file mới với lệnh DROP TABLE IF EXISTS: {output_path}')
```
