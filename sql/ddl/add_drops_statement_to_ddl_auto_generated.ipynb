```python
ddl_schema = re.sub(
    r'(CREATE SCHEMA\s+[`"\[]?[a-zA-Z0-9_]+[`"\]]?)',
    lambda m: f'DROP SCHEMA IF EXISTS {m.group(0)[13:]};\n{m.group(0)}',
    ddl
)
```

```python
import re

def add_drop_table(match):
    full_table = match.group(1)  # e.g. `dim`.`customers` or `customers`
    # Extract table name after last dot (.) and remove backticks if any
    table_name = re.search(r'`?([a-zA-Z0-9_]+)`?$', full_table).group(1)
    return f'DROP TABLE IF EXISTS `{table_name}`;\nCREATE TABLE {full_table}'


ddl_new = re.sub(
    r'CREATE TABLE\s+([`"\[\]a-zA-Z0-9_.]+)',
    add_drop_table,
    ddl
)

# with open('output.sql', 'w', encoding='utf-8') as f:
#     f.write(ddl_new)
ddl_new
```
